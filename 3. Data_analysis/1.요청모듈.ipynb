```python
# ÏùëÎãµÏΩîÎìúÎ•º ÌôúÏö©Ìïú Ï°∞Í±¥Î∂ÑÍ∏∞
def url_check(url):
    res = rq.get(url)
    
    print(res)
    
    sc = res.status_code
    
    if sc == 200:
        print(f"{url} ÏöîÏ≤≠ ÏÑ±Í≥µ")
        
    elif sc == 404:
        print(f"{url} Ï∞æÏùÑ Ïàò ÏóÜÏùå")
        
    else:
        print(f"{url} Ïïå Ïàò ÏóÜÎäî ÏóêÎü¨ : {sc}")
```

```python
url_check("https://www.naver.com/")
```

```python
url_check("https://www.naver.com/a")
```
